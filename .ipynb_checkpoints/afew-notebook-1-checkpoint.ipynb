# Advanced Fisheries Economics Workshop

This is a Jupyter note book. You can write fromatted text in a Jupyter notebook using markdown, you can do mathematics, you can analyze data and you can plot graphs. You can import Excel files 

Notebooks or interactie notebooks are a way of working that combines text and computation, they are free, don't take up a lot of space (lightweight) and are portable. They can be e-mailed attached in a Slack message, shared over the internet via [Google collaboaratory](https://colab.research.google.com/) or [Microsoft Azure notebooks](https://notebooks.azure.com/) or any number of a range of other platforms.

They can also be used to do presentations (slideshows) with the addition of package called [RISE](https://github.com/damianavila/RISE).
