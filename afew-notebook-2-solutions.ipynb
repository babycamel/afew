```python
#import handcalcs.render
import math as math # mathematical functions
from sympy import * #symbolic algebra package
import numpy as np # numerical python packagee
import matplotlib.pyplot as plt #matplotlib is a plotting package
%matplotlib inline  
# without this plots are popouts
```

```python
# constants
x0 = 1 #choose a starting value
r = 0.2 #choose a growth rate
t = np.linspace(0,10,20) #define an x-axis
x = x0*np.exp(r*t) #function to plot y-axis
# instructions for plotting
plt.plot(t,x) # plot pairs of values
plt.xlabel("t") # creat an x label
plt.ylabel("x(t)") #create a y-label
plt.title("Exponential growth")
```

```python
x0 = 1 #choose a starting value
r = 0.2 #choose a growth rate
K = 100
t = np.linspace(0,100,100) #define an x-axis uses numpy
#x = x0*np.exp(r*t) #function to plot y-axis CHANGE THIS LINE
x = K*x0*np.exp(r*t)/(K + x0*np.exp(r*t) - x0) #logistic
plt.plot(t,x) # plot pairs of values

#below here is decoration
plt.xlabel("time")
plt.ylabel("stock of fish")
plt.title("Logistic growth")

```

```python
#Constants 
r = 2.4 # intrinsic growth rate
K = 100 # carrying capacity/unfished biomass
x0 = 1 #initial value
#initialize values
b  = 100 # excel number of rows
x = np.zeros(b)  # vector of zeroes 
#print(x)
x[0] = x0 # pass initial value to vector
for i in range(0,len(x)-1):  #for loop
     x[i+1] = x[i] + r*x[i]*(1-x[i]/K)
        
print(x)
```
